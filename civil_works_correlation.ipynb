```python
import pandas as pd
import numpy as np

# notices_df = pd.read_csv('datasets/PhilGEPS Bids and Awards Notices (Province of South Cotabato) - 2011 to 2019.09.csv')
# notices_df = notices_df[pd.notnull(notices_df['Awardee Corporate Title'])]
# notices_df.replace(np.nan, -99999, inplace=True)
# notices_df.rename(columns={'Awardee Corporate Title':'Contractor / Supplier'}, inplace=True)

civil_df = pd.read_csv('datasets/Procurement Monitoring Report (Civil Works) - 2018.csv', low_memory=False)
civil_df = civil_df[pd.notnull(civil_df['Contractor / Supplier'])]
civil_df.replace(np.nan, -99999, inplace=True)
civil_df.rename(columns={'Mode of Procurement':'Procurement Mode',
                         'Approved Budget of Contract (ABC)':'ABC',
                         'Contract Cost':'Cost'}, inplace=True)

#REFERENTIAL PURPOSE
# print('PhilGEPS Bids and Awards Notices \n(Province of South Cotabato)\n')
# for col in notices_df.columns:
#     print(col)
    
# print('\n\nProcurement Monitoring Report (Civil Works)\n')
# for col in civil_df.columns:
#     print(col)
```

```python

ace = []
procurement_mode_list = []
contractor_list = []
civil_df['Cost'] = civil_df['Cost'].replace(',','',regex=True).astype(float)
civil_df['ABC'] = civil_df['ABC'].replace(',','',regex=True).astype(float)
civil_df['Cost Efficiency'] = civil_df.apply(lambda row: ((row.Cost / row.ABC)*100), axis=1)

for contractor, group_of_contractor in civil_df.groupby('Contractor / Supplier'):
    for procurement_mode, grouped_group_of_contractor in group_of_contractor.groupby('Procurement Mode'):
        #print(grouped_group_of_contractor['Cost Efficiency'].mean(), procurement_mode, contractor)
        ace.append(grouped_group_of_contractor['Cost Efficiency'].mean())
        procurement_mode_list.append(procurement_mode)
        contractor_list.append(contractor)
act_mode_df = pd.DataFrame({'Ave. Cost Efficiency':ace, 'Contractor':contractor_list, 'Procurement Mode': procurement_mode_list})
# mode_df = pd.DataFrame()
# for contractor in act_mode_df['Contractor'].values[:5]:
#     print(contractor)
#     act_mode_df = mode_df.join(act_mode_df.set_index('Procurement Mode')[['Ave. Cost Efficiency']].rename(columns={'Ave. Cost Efficiency':contractor}))
#        act_mode_df.assign(Average_Cost_Efficiency=grouped_group_of_contractor['Cost Efficiency'].mean())
#        print(grouped_group_of_contractor['Cost Efficiency'].mean() ,procurement_mode,contractor)
#         for i in range(10):
#             if act_mode_df.empty:
#                 act_mode_df = grouped_group_of_contractor.set_index('Procurement Mode')[['Cost Efficiency']].rename(columns={'Cost Efficiency':contractor})
#             elif not act_mode_df.empty:
#                 act_mode_df = act_mode_df.join(grouped_group_of_contractor.set_index('Procurement Mode')[['Cost Efficiency']].rename(columns={'Cost Efficiency':contractor}))
                
```
