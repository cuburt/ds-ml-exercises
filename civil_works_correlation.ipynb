```python
import pandas as pd
import numpy as np

notices_df = pd.read_csv('datasets/PhilGEPS Bids and Awards Notices (Province of South Cotabato) - 2011 to 2019.09.csv')
notices_df = notices_df[pd.notnull(notices_df['Awardee Corporate Title'])]
notices_df.replace(np.nan, -99999, inplace=True)
notices_df.rename(columns={'Awardee Corporate Title':'Contractor / Supplier'}, inplace=True)


civil_df = pd.read_csv('datasets/Procurement Monitoring Report (Civil Works) - 2018.csv')
civil_df = civil_df[pd.notnull(civil_df['Contractor / Supplier'])]
civil_df.replace(np.nan, -99999, inplace=True)
civil_df.rename(columns={'Mode of Procurement':'Procurement Mode',
                         'Approved Budget of Contract (ABC)':'ABC',
                         'Contract Cost':'Cost'}, inplace=True)

#REFERENTIAL PURPOSE
print('PhilGEPS Bids and Awards Notices \n(Province of South Cotabato)\n')
for col in notices_df.columns:
    print(col)
    
print('\n\nProcurement Monitoring Report (Civil Works)\n')
for col in civil_df.columns:
    print(col)
```

```python


act_mode_df = pd.DataFrame()

civil_df['Cost'] = civil_df['Cost'].replace(',','',regex=True).astype(float)
civil_df['ABC'] = civil_df['ABC'].replace(',','',regex=True).astype(float)
civil_df['Cost Efficiency'] = civil_df.apply(lambda row: (row.Cost / row.ABC), axis=1)
for name, group in civil_df.groupby('Contractor / Supplier'):
    if act_mode_df.empty:
        act_mode_df = group.set_index('Procurement Mode')[['Cost Efficiency']].rename(columns={'Cost Efficiency':name})
    else: 
        act_mode_df = act_mode_df.join(group.set_index('Procurement Mode')[['Cost Efficiency']].rename(columns={'Cost Efficiency':name}))
```
