```python
#CUBE CONSOLE
class cube:
    
    def __init__(self, length):
        self.length = float(length)
        
    def cubevol(self):
        return   (float(self.length**3))
    
    def squarearea(self):
        return   (float((self.length)*(self.length)))
    
    def cubearea(self):
        return   ((self.length * self.length)*6)/2

    def cube(self):
        return self.length
    
cubic = cube(input())
print("The volume of cube is: %s" % (cubic.cubevol()))
print("The area of square is: %s" % (cubic.squarearea()))
print("The area of cube is: %s" % (cubic.cubearea()))
print(cubic.cube())
```

```python
from tkinter import *
import math



class Calculate:
    def __init__(self, master):
        self.master = master
        master.title('Python Volume Calculator')
        height_lbl = Label(master, text="Height").grid(row=0, column=0)
        radius_lbl = Label(master, text="Radius").grid(row=1, column=0)
        
        height = Entry(master)
        height.grid(row=0, column=1)
        
        radius = Entry(master)
        radius.grid(row=1, column=1)
        
        cone = Button(master, text="cone", command=lambda: self.cone_vol(radius, height)).grid(row=4, column=3)
        
        cylinder = Button(master, text="cylinder", command=lambda: self.cylinder_vol(radius, height)).grid(row=4, column=4)
        
        sphere = Button(master, text="sphere", command=lambda: self.sphere_vol(radius)).grid(row=4, column=5)
        
    def sphere_vol(self, rad):
        self.get_answer(4/3*math.pi*(float(rad.get())**3), 'sphere')
        
    def cylinder_vol(self, rad, hgt):
        self.get_answer(math.pi*(float(rad.get())**2)*(float(hgt.get())), 'cylinder')

    def cone_vol(self, rad, hgt):
        self.get_answer(math.pi*(float(rad.get())**2)*(float(hgt.get()))/3, 'cone')
        
    def get_answer(self, answer, shape):
        answer_lbl = Label(self.master, text='The volume of the %s is \n %.2f cubic meters.'%(shape ,round(answer,2)), fg='red').grid(row=0, column=2, rowspan=2, columnspan=4)

root = Tk()
gui = Calculate(root)
root.mainloop()
```
